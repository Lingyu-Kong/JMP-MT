```python
from __future__ import annotations

from fairchem.core.datasets import AseDBDataset

dataset_path = "/storage/nima/omat24/val/rattled-1000/"
config_kwargs = {}  # see tutorial on additional configuration

dataset = AseDBDataset(
    config=dict(
        src=dataset_path,
        a2g_args={
            "r_energy": True,
            "r_forces": True,
            "r_stress": True,
        },
        **config_kwargs,
    )
)
```

```python
import nshutils as nu

# atoms objects can be retrieved by index
atoms = dataset.get_atoms(0)
nu.display(atoms.get_stress(voigt=False))
```

```python
nu.display(dataset[0].to_dict())
```
